```
df.drop(columns=['Building name-1', 'Images',], inplace=True)
```

```

df['City'] = df['City'].str.strip().str.lower()
df['Address'] = df['Address'].str.strip().str.title()
df['PossesionDate'] = pd.to_datetime(df['PossesionDate'], errors='coerce')
```

```
df['price'] = df['price'].astype(str).str.extract(r'(\d+\.\d+|\d+)').astype(float)
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('-', '_')
df['CarpetArea'] = df['CarpetArea'].astype(str).str.extract(r'(\d+)').astype(float)
df['Floors'] = df['Floors'].astype(str).str.extract(r'(\d+)').astype(float)
df['AgeOfProperty'] = df['AgeOfProperty'].astype(str).str.extract(r'(\d+)').astype(float)
```

```
df.isnull().sum()
```

```
df['Amenities'].fillna('Unknown', inplace=True)
df['Facing'].fillna('Unknown', inplace=True)
df['bathroom'].fillna(df['bathroom'].median(), inplace=True)
df['parking'].fillna(0, inplace=True)
df['Buildingname'].fillna('Unknown', inplace=True)
df['CarpetArea'].fillna(df['CarpetArea'].median(), inplace=True)
df['Floors'].fillna(df['Floors'].median(), inplace=True)
df['AgeOfProperty'].fillna(df['AgeOfProperty'].median(), inplace=True)
df['City'].fillna('Unknown', inplace=True)
df['Address'].fillna('Unknown', inplace=True)
df['PossesionDate'].fillna('Unknow', inplace=True)
```
